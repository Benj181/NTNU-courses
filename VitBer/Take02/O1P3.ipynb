```python
import numpy as np
import matplotlib.pyplot as plt
from functools import reduce
import sympy 
```

```python
def lagrange_polys(x, xs):
    n = len(xs)
    return [[reduce(lambda x,y:x*y, [(x - xs[j])/(xs[i] - xs[j]) for j in range(n) if j != i])] for i in range(n)]

def polys(ls, ys):
    return reduce(lambda x,y:x+y, np.array(ys)@np.array(ls))
```

```python
tdata = np.array([0 , 0.8976, 1.7952, 2.6928, 3.5904 , 4.4880 , 5.3856 , 6.2832])
xdata = np.array([1 , 1.5984, -0.6564 , -1.6828 , -0.1191 ,0.2114, -0.3514 , 1 ])
ydata = np.array([0 , 0.7818, 0.9750, 0.4339, -0.4339, -0.975, -0.7818, 0])
```

```python
xf = lambda t: np.cos(t) + np.sin(2*t)
yf = lambda t: np.sin(t)
```

```python
t = sympy.symbols('t')

ls = lagrange_polys(t, tdata)
ps = polys(ls, xdata)
px = sympy.lambdify(t, ps)

ls = lagrange_polys(t, tdata)
ps = polys(ls, ydata)
py = sympy.lambdify(t, ps)

ts = np.linspace(tdata[0], tdata[-1], 1000)

plt.plot(tdata, xdata, "o")
plt.plot(ts, px(ts))
plt.plot(ts, xf(ts))
plt.grid()
plt.show()

plt.plot(tdata, ydata, "o")
plt.plot(ts, py(ts))
plt.plot(ts, yf(ts))
plt.grid()
plt.show()
```
